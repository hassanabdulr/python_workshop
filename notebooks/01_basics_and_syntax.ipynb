## 1. Basic Arithmetic Operations

Python supports standard arithmetic operations:

```python
# Addition
result_add = 10 + 5
print(f"10 + 5 = {result_add}")

# Subtraction
result_sub = 10 - 5
print(f"10 - 5 = {result_sub}")

# Multiplication
result_mult = 10 * 5
print(f"10 * 5 = {result_mult}")

# Division (returns a float)
result_div = 10 / 3
print(f"10 / 3 = {result_div}")
```

## 2. Floor Division and Modulo

Two important operations for working with integers:

```python
# Floor division (//) - divides and rounds down to nearest integer
floor_div = 10 // 3
print(f"10 // 3 = {floor_div}")

# Modulo (%) - returns the remainder of division
modulo = 10 % 3
print(f"10 % 3 = {modulo}")

# Practical example: checking if a number is even or odd
number = 42
if number % 2 == 0:
    print(f"{number} is even")
else:
    print(f"{number} is odd")
```

## 3. Variable Assignment

Variables store values for later use:

```python
# Assigning values to variables
x = 10
y = 5
name = "Python"
pi = 3.14159

print(f"x = {x}")
print(f"y = {y}")
print(f"name = {name}")
print(f"pi = {pi}")

# Variables can be used in operations
result = x + y
print(f"x + y = {result}")

# Multiple assignment
a, b, c = 1, 2, 3
print(f"a={a}, b={b}, c={c}")
```

## 4. String Operations

Strings are sequences of characters and support various operations:

```python
# String concatenation
first_name = "Python"
last_name = "Workshop"
full_name = first_name + " " + last_name
print(full_name)

# String repetition
repeat = "Ha" * 3
print(repeat)

# String length
message = "Hello, World!"
print(f"Length of '{message}': {len(message)}")

# String indexing (0-based)
print(f"First character: {message[0]}")
print(f"Last character: {message[-1]}")

# String slicing
print(f"First 5 characters: {message[0:5]}")
print(f"From index 7 to end: {message[7:]}")
```

## 5. String Methods

Python strings have many useful built-in methods:

```python
text = "  Introduction to Python  "

# Convert to uppercase
print(text.upper())

# Convert to lowercase
print(text.lower())

# Remove leading/trailing whitespace
print(text.strip())

# Replace substring
print(text.replace("Python", "Programming"))

# Split string into list
words = "apple,banana,cherry".split(",")
print(words)

# Join list into string
joined = " - ".join(words)
print(joined)
```
